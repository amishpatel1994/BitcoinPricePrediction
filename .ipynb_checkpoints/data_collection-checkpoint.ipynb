## Daily Bitcoin Stats Summary

These are the properties we will be using for the price prediction

~~1. MWNUS -> Bitcoin My Wallet Number of Users~~
2. MKPRU -> Bitcoin Market Price USD.
4. DIFF -> Bitcoin Difficulty. This is the measure of how difficult it is to find a new block. That is, this is the minimum proof-of-work threshold required for the miner to mine a block.
~~5. MWNTD -> Bitcoin My Wallet Number of Transactions Per Day
The number of daily confirmed Bitcoin transactions~~
~~6. MWTRV -> My Wallet Transaction Volume~~
7. AVBLS -> Bitcoin Average Block Size.
Storage size of each block
~~8. BLCHS -> Bitcoin api.blockchain Size
The storage size of the blockchain~~
9. MIREV -> Bitcoin Miners Revenue.
Total value of coinbase block rewards and transaction fees paid to miners.
10. HRATE -> Bitcoin Hash Rate.
This is the estimated number of tera hashes per second the bitcoin network is performing.
11. CPTRA -> Bitcoin Cost per Transaction.
This is the miners revenue divided by the number of transactions
15. NTRBL -> Bitcoin Number of Transaction per Block.
This is the number of transactions in each block
16. NADDU -> Bitcoin Number of Unique Bitcoin Addresses Used.
The total number of unique addresses used on the Bitcoin Blockchain for the day
~~17. NTRAT -> Bitcoin Total Number of Transactions
Total number of transactions to date.~~
18. NTRAN -> Bitcoin Number of Transactions.
Total number of unique bitcoin transactions per day.
19. TRFEE -> Bitcoin Total Transaction Fees.
Total value of all transaction fees paid to miners in BTC for the day
~~20. MKTCP -> Bitcoin Market Capitalization.~~
~~21. TOTBC -> Total Bitcoins.~~

```python
# all_blockchain_props = ['MWNUS', 'MKPRU', 'TRVOU', 'DIFF', 'MWNTD', 'MWTRV', 'AVBLS', 'BLCHS', 'MIREV', \
#                     'HRATE', 'CPTRA', 'CPTRV', 'ETRVU', 'TOUTV', 'NTRBL', 'NADDU', 'NTRAT', 'NTRAN', \
#                     'TRFEE', 'MKTCP', 'TOTBC']

blockchain_props = ['MKPRU', 'DIFF', 'AVBLS', 'MIREV', 'HRATE', 'CPTRA', 'NTRBL', 'NADDU', 'NTRAN', 'TRFEE']

data = None
quandl.ApiConfig.api_key = 'APIKEY'
for prop in blockchain_props:
    bc_prop = 'BCHAIN/' + prop
    if data is None:
        data = quandl.get(bc_prop, start_date='2014-12-18')
    else:
        data = pd.merge(data, quandl.get(bc_prop, start_date='2014-12-18'), on='Date')

data.columns = ['Market Price', 'Difficulty', 'Block Size', 'Miner Revenue Value', 'Hash Rate', 'Transaction Amount', 'Transactions per Block', 'Unique Addresses', 'Number of Transactions', 'Transaction Fees']
data.tail()
```
